```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
```

#### Function

This is a general wavefunction for a traveling wave in the x-direction, which returns the displacement (y) value. 

```python
def Y(x,t,A,k,omega):
    return A*np.cos(k*x - omega*t)
```

**Define the parameters for the wave.** 

We get omega from the wavenumber (k) values, with c = omega / k. We set the speed of light c to be 1, since we are not giving physical units.

We will create seven waves, which range in k from 5$\pi$ to 8$\pi$ (in steps of 0.5$\pi$). The amplitudes A of the waves follow roughly (very roughly) a Gaussian shape, which we will plot below.

```python
def Ytime(x,A,k,omega):
    return lambda t:Y(x,t,A,k,omega)

#Ytime returns a wave with given parameters as a function of time

waves = [Ytime(3,A,k,k) for k,A in zip([np.pi * (5 + i/2) for i in range(0,7)], [1/4,1/3,1/2,1,1/2,1/3,1/4])]

print([wave(1) for wave in waves])
```

#### Set the x and t values for the waves

In this case, we are viewing the waves only at x = 3.
The t values range from 0 to 6.

#### Create the wave packet

Here we use the function to create the y values for each wave (at each x value), and we add the y values of the five waves (at each x value).

#### Make the plot of the individal waves

Note, an artificial offset is given to six of the waves so they are easier to see:

```python
[plt.plot(np.linspace(0,2,500), [wave(i) + 1.6 * j for i in np.linspace(0,2,500)]) for j, wave in enumerate(waves)]
plt.xlabel("time")
plt.show()
```

#### Make the plot of the wave packet


```python
plt.plot(np.linspace(0,6,500),[sum([wave(time) for wave in waves]) for time in np.linspace(0,6,500)])
plt.xlabel("time")
plt.ylabel("amplitude")
plt.show()
```

#### Make the plot of the wave amplitudes as a function of k

```python
plt.plot([np.pi * (5 + i/2) for i in range(0,7)],[1/4,1/3,1/2,1,1/2,1/3,1/4])
plt.xlabel("wave number")
plt.ylabel("amplitude")
```
